# Import the "qrcode" module

```python
import qrcode
from PIL import Image
import matplotlib.pyplot as plt
```

# Create a QR Code Instance

```python
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
```

# Set the data you want to encode in the QR Codde

```python
data = "https://github.com/"
qr.add_data(data)
qr.make(fit=True)
```

# Create an image of the QR Code with Specified colors

```python
img = qr.make_image(fill_color="black", back_color="white")
```

# Display Image in google Collab

```python
plt.imshow(img, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```

# Save the image

```python
img.save("qrcode.png")
print("QR code saved as 'qrcode.png'")

# to view the image search "qrcode.png" in search bar or QR code will be in the same directory where your Python script is located
```
